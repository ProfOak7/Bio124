```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import sem
```

```python
df = pd.read_excel (r'./HR_ex.xlsx')
df
```

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, x = 'Group', y = 'Favg',color = 'r', marker = 'o', yerr = 'Fsem', capsize=5)
df.plot.line(ax = ax, x = 'Group', y = 'Mavg',color = 'b', marker = 'o', yerr = 'Msem', capsize=5)
ax.set_xlabel("Time (min)")
ax.set_ylabel("HR (bpm)")
ax.set_xticklabels(ax.get_xticklabels(), rotation=25)
```

```python
F_threshold = 63
M_threshold = 68.666667
```

```python
df['N_Favg'] = df.Favg/63
df
```

```python
df['N_Mavg'] = df.Mavg/68.666667
df
```

```python
df['N_Fsem'] = sem(df.N_Favg)
df
```

```python
df['N_Msem'] = sem(df.N_Mavg)
df
```

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, x = 'Group', y = 'N_Favg',color = 'r', marker = 'o', yerr = 'N_Fsem', capsize=5)
df.plot.line(ax = ax, x = 'Group', y = 'N_Mavg',color = 'b', marker = 'o', yerr = 'N_Msem', capsize=5)
ax.set_xlabel("Time (min)")
ax.set_ylabel("HR (Normalized to Resting HR)")
ax.set_xticklabels(ax.get_xticklabels(), rotation=25)
```

```python
threshold = 1
```

```python
# Exclude the first two rows and then check for the condition
reached_threshold_M = df['N_Mavg'].iloc[2:] <= threshold
time_at_threshold_M = df.loc[reached_threshold_M.index[reached_threshold_M], 'Group'].iloc[0] if reached_threshold_M.any() else None

reached_threshold_F = df['N_Favg'].iloc[2:] <= threshold
time_at_threshold_F = df.loc[reached_threshold_F.index[reached_threshold_F], 'Group'].iloc[0] if reached_threshold_F.any() else None

print(f"Time to reach ~resting HR (set to: {threshold}) for the males was: {time_at_threshold_M} minutes")
print(f"Time to reach ~resting HR (set to: {threshold}) for the females was: {time_at_threshold_F} minutes")
```
