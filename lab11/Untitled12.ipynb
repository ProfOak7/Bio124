```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel (r'./HR_ex.xlsx', index_col=0)
df
```

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, y = 'Favg',color = 'r', marker = 'o', yerr = 'Fsem', capsize=5)
df.plot.line(ax = ax, y = 'Mavg',color = 'b', marker = 'o', yerr = 'Msem', capsize=5)
ax.set_xlabel("Time (min)")
ax.set_ylabel("HR (bpm)")
```
